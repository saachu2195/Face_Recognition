```python
## STEP 1:

import cv2
import os
os.chdir("C:/Users/square/Desktop/Face_Recognition_Detector")
import numpy as np

## Importing the Haar Cascade
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

## Adding a video capture object
cap = cv2.VideoCapture(0)
id = input("Please Enter your User_ID: ")
sampleNum = 0
while(True):
    ## Capture frame by frame
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray, scaleFactor = 1.4, minNeighbors = 5)
    print("Number of Faces", len(faces))
    
    for (x, y, w, h) in faces:         ## X & y are the location of rectangle.
        sampleNum = sampleNum+1
        roi_gray = img[y:y+h, x:x+w]
        cv2.imwrite("dataset/users." + str(id)+ "." + str(sampleNum) + ".jpg", roi_gray)
        print(x, y, w, h)              ## w & h are the width & height.
        
        ## Drawing a Rectangle
        color = (200, 100, 50)
        stroke = 2
        width = x+w
        height = y+h
        cv2.rectangle(img, (x, y), (width, height), color, stroke)
        cv2.waitKey(100)
        
    ## Display the resulting frame
    cv2.imshow('frame', img)
    cv2.waitKey(1)
    if(sampleNum > 15):
        break
        
## When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
## Getting the path.

import cv2
from PIL import Image
import numpy as np
import os

os.chdir("C:/Users/square/Desktop/Face_Recognition_Detector")

## Getting the path.

path = "C:/Users/square/Desktop/Face_Recognition_Detector/dataset"
recognizer = cv2.face.LBPHFaceRecognizer_create()


def getImagesPath(path):
    imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
    print(imagePaths)
    
    faces = []
    Ids= []
    
    for imagepath in imagePaths:
        faceImg = Image.open(imagepath).convert('L')
        facearray = np.array(faceImg)
        Id = int(os.path.split(imagepath)[-1].split('.')[1])
        faces.append(facearray)
        print(Id)
        Ids.append(Id)
        cv2.imshow("training_data", facearray)
        cv2.waitKey(10)
        
    return Ids, faces

ids, faces = getImagesPath(path)
recognizer.train(faces, np.array(ids))
recognizer.save('recognizer/trainingData.yml')
cv2.destroyAllWindows()
```

```python
import os
os.chdir("C:/Users/square/Desktop/Face_Recognition_Detector")
import numpy as np
import cv2

face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cam = cv2.VideoCapture(0)

##Uploading the data
rec = cv2.face_LBPHFaceRecognizer.create()
rec.read("C:/Users/square/Desktop/Face_Recognition_Detector/recognizer/trainingData.yml")
id = 0
font = cv2.FONT_HERSHEY_PLAIN

while(True):
    ## Capture frame by frame
    ret, img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray, scaleFactor = 1.4, minNeighbors = 5, minSize = (30, 30))
    print("Number of Faces", len(faces))
    for (x, y, w, h) in faces:         ## X & y are the location of rectangle.
        print(x, y, w, h)              ## w & h are the width & height.
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        
        ## Drawing a Rectangle
        color = (200, 100, 50)
        stroke = 2
        width = x+w
        height = y+h
        cv2.rectangle(img, (x, y), (width, height), color, stroke)
        id, conf = rec.predict(gray[y:y+h, x:x+w])
        if (id==1):
            id="SACHIN {0:.2f}%".format(round(100 - conf, 2))
        elif(id==2):
            id="MOM {0:.2f}%".format(round(100-conf, 2))
        elif(id==3):
            id="Duregesh {0:.2f}%".format(round(100-conf, 2))
        elif(id==4):
            id='Suraj'
        cv2.putText(img, str(id), (x, y+h), font, color=(0, 0, 255), fontScale=2)
        
    ## Display the resulting frame
    cv2.imshow('Faces', img)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
## When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
